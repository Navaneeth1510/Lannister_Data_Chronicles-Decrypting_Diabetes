# Dataset Visualization using Heatmaps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "C:/Users/NARENDRA/Downloads/The Lannister Data Chronicles_ Decrypting Diabetes.csv"

data_frame = pd.read_csv(file_path)
```

```python
# heatmap vis for diabetes = 0


df = data_frame[data_frame['Diabetes_012'] == 0]
df = df.drop(columns=['Diabetes_012'])
#finding the correlations
correlation_matrix = df.corr()

#plotting the heatmap
plt.figure(figsize=(14, 12))

heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

heatmap.set_title('Correlation Matrix of Diabetes = 0 Data', fontsize=16)
plt.show()

print("----------------------------------------------------------------------------------------------------------------------")
```
